```python
import pytesseract
from PIL import Image, ImageFilter, ImageOps
import cv2
import numpy as np

# Caminho da imagem
image_path = "/home/janderson/Documentos/challenge-artificial-intelligence/resources/Infografico-1.jpg"

# --- 1️⃣ Ler imagem com OpenCV ---
img_cv = cv2.imread(image_path)

# --- 2️⃣ Converter para escala de cinza ---
gray = cv2.cvtColor(img_cv, cv2.COLOR_BGR2GRAY)

# --- 3️⃣ Remover ruído e aumentar contraste ---
gray = cv2.medianBlur(gray, 3)
gray = cv2.equalizeHist(gray)

# --- 4️⃣ Binarizar (preto e branco) ---
_, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# --- 5️⃣ Converter novamente para PIL (pytesseract usa PIL) ---
processed_image = Image.fromarray(thresh)

# --- 6️⃣ Extrair texto com Tesseract ---
custom_config = r'--oem 3 --psm 6'  # modo LSTM e layout semi-estruturado
extracted_text = pytesseract.image_to_string(processed_image, lang="por", config=custom_config)

print("=== TEXTO EXTRAÍDO ===")
print(extracted_text)
```
